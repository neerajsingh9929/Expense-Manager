```python
import pandas as pd
DF=pd.read_excel(uploaded['plot price dataset.xlsx'])
```

```python
DF
y=DF['plot Price']
x=DF['plot size']
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
import numpy as np

def fit(x,y,epoch,lr):
  m=0
  c=0
  (n,)=np.shape(y)
  for j in range(epoch):

     for i in  range(n):
        e=(m*x[i]+c)-y[i]
        gm=-(2*e*x[i])/n
        gc=-(2*e)/n
        m=m+lr*gm
        c=c+lr*gc
  return (m,c)
```

```python
def predict(x,m,c):
  (n,)=np.shape(x)
  p=np.zeros(n)
  for i in range(n):
    p[i]=m*x[i]+c
  return p
```

```python
(m1,c1)=fit(x,y,100,0.001)
```

```python
pred=predict(x,m1,c1)
```

```python
plt.scatter(x,y)
plt.plot(x,pred)
```

#now with sk-learn


```python
from sklearn .linear_model import LinearRegression as lr
LR=lr()
(n,)=np.shape(x)
x2=x.to_numpy()
y2=y.to_numpy()
x3=x2.reshape(n,1)
y3=y2.reshape(n,1)

LR.fit(x3,y3)
pred=LR.predict(x3) 
plt.scatter(x,y)
plt.plot(x,pred)
print (LR.score(x3,y3))
```
